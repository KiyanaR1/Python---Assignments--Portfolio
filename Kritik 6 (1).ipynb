```python


import numpy as np
import math

def normal_density (mean, variance, x):

    # Calculate the probability density function (PDF)
    pdf = (1 / (np.sqrt(2 * np.pi *variance)))* np.exp(-((x - mean) ** 2) / (2 * variance))

    return pdf

def integration(mean, variance, a, b, n ):
    
    delta_x = (b - a) / n ### determines the width of the trapazoids
    int_value = 0 ### initializes sum values
    
    for i in range(1, n+1):
        x_i = a + (i-1) * delta_x
        int_value += (normal_density (mean,variance, x_i) + normal_density (mean,variance, a + i * delta_x)) * delta_x / 2 ## adds up areas of trapozoids under the curve up together
        
        

    return int_value 

mean_value = 171
variance_value = 7.1**2
lower_limit = 162
upper_limit = 190
num_intervals = 1000

result = integration(mean_value, variance_value, lower_limit, upper_limit, num_intervals)
print("Approximated Integral:", result)
```
